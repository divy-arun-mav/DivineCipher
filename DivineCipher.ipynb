```python
import numpy as np

def encryption(plain_text):
    ALPHABETS = [chr(i) for i in range(ord('A'), ord('Z')+1)]
    
    key1 = sum(ord(char) for char in plain_text) % 26
    print("Key k1: ",key1)
    
    k2, k3 = 7, 2
    
    P=[(ALPHABETS.index(char)) for char in plain_text]
    
    p1 = [(ALPHABETS.index(char) ^ len(plain_text)) % 26 for char in plain_text]
    

    p2 = [(p_val + key1) % 26 for p_val in p1]
    
    matrix_size = int(np.ceil(len(p2) ** 0.5))
    p3 = np.array(p2 + [0] * (matrix_size ** 2 - len(p2))).reshape(matrix_size, matrix_size)
    p3 = np.rot90(p3, k=-1)                                                                      

    p4 = np.array([((p_val ^ k3) * k2) % 26 for p_val in p3.flatten()])
    
    # Ensure values remain within range 0 to 25
    p4 = p4 % 26
    
    num_zeros = matrix_size ** 2 - len(p4)
    p4_with_zeros = np.append(p4, np.zeros(num_zeros))
    
    # Convert p4_with_zeros to integer dtype
    p4_with_zeros = p4_with_zeros.astype(int)
    print(p4)
    
    subarrays = [p4_with_zeros[i:i+3] for i in range(0, len(p4_with_zeros), 3)]
    for i, subarray in enumerate(subarrays):
        if len(subarray) < 3:
            subarrays[i] = np.append(subarray, [ord(special_char) for special_char in 'AB'])
    
    for i, subarray in enumerate(subarrays):
        subarrays[i][[0, 2]] = subarrays[i][[2, 0]]
    
    merged_array = np.concatenate(subarrays)
    
    merged_array = [(num+k3) % 26 for num in merged_array]
    print(merged_array)
    
    cipher_text = ''.join(ALPHABETS[p_val] for p_val in merged_array)
    
    return cipher_text

plain_text = 'DJSCOE'
cipher_text = encryption(plain_text)
print("Plaintext:", plain_text)
print("Encrypted Data:", cipher_text)
```
